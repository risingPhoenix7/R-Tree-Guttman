```python
import sys
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import random
import re

def read_points_file(file_path):
    with open(file_path) as file:
        points = [tuple(map(int, re.findall(r'\d+', line))) for line in file]
    return points

def read_rectangles_file(file_path):
    with open(file_path) as file:
        rectangles = [tuple(map(int, re.findall(r'\d+', line))) for line in file]
        rectangles = [((rectangles[i][0], rectangles[i][1]), (rectangles[i][2], rectangles[i][3])) for i in range(len(rectangles))]
    return rectangles


def draw_points_and_rectangles(points, rectangles,file_index,label):
    fig, ax = plt.subplots(figsize=(10, 10))

    for point in points:
        ax.scatter(*point, c='blue', marker='o', s=50)
        ax.text(point[0] + 0.1, point[1] + 0.1, f'({point[0]}, {point[1]})', fontsize=8)

    # Sort rectangles by area
    rectangles.sort(key=lambda r: (r[1][0] - r[0][0]) * (r[1][1] - r[0][1]))

    for rectangle in rectangles:
        bottom_left, top_right = rectangle
        width = top_right[0] - bottom_left[0]
        height = top_right[1] - bottom_left[1]
        color = (random.random(), random.random(), random.random())

        if width == 0 or height == 0:
            # Draw a line instead of a rectangle
            ax.plot([bottom_left[0], top_right[0]], [bottom_left[1], top_right[1]], color=color)
        else:
            ax.add_patch(patches.Rectangle(bottom_left, width, height, linewidth=1, edgecolor=color, facecolor=color, alpha=0.1))

    min_x = min(min(p[0] for p in points), min(r[0][0] for r in rectangles))
    max_x = max(max(p[0] for p in points), max(r[1][0] for r in rectangles))
    min_y = min(min(p[1] for p in points), min(r[0][1] for r in rectangles))
    max_y = max(max(p[1] for p in points), max(r[1][1] for r in rectangles))

    ax.set_xlim(min_x - 1, max_x + 1)
    ax.set_ylim(min_y - 1, max_y + 1)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_aspect('equal', adjustable='box')

    ax.set_title(label)
    plt.savefig(f"graph_{file_index}{label}.png")
    plt.show()
    plt.close(fig)


def main():
    points_file = 'datasmall.txt'  # Replace with your points file name
    points = read_points_file(points_file)

    for i in range(1, 22):  # Assuming files are numbered 1 to 21
        rectangles_file = f'printrectangles{i}.txt'
        rectangles = read_rectangles_file(rectangles_file)
        label = f"insert point{points[i-1]}"
        
        # Take the points up to the current index (i)
        points_until_index = points[:i]

        # Draw and save the graph
        draw_points_and_rectangles(points_until_index, rectangles, i, label)




if __name__ == '__main__':
    main()
```
